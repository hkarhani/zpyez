# Zero PyEZ Testing from NodeJS Client 

## 1. Initialize connection to ZPyEZ Container 

```nodejs
var zerorpc = require("zerorpc");

var client = new zerorpc.Client();
client.connect("tcp://192.168.99.100:8585");
```

```nodejs
_dev1 = {host: '172.30.96.56', user:'user', password:'secret'}
```

```nodejs
_dev2 = {host: '172.30.96.3', user:'user', password:'secret'}
```

## 2. Opening PyEZ NetConf/ssh Sessions from NodeJS

```nodejs
client.invoke("open", _dev1, function(error, res) {
    if (error){
        console.log('Error: ', error);
    }
    dev1 = res.toString(); 
    console.log(res.toString());
});
```

```nodejs
client.invoke("open", _dev2, function(error, res) {
    if (error){
        console.log('Error: ', error);
    }
    dev2 = res; 
    console.log(res.toString());
});
```
